```python
import cv2
import matplotlib.pyplot as plt

from utils import load_image, display_image
```

```python
image = load_image("photo.jpg")
```

```python
display_image(image)
```

```python
imageLine = image.copy()
cv2.line(imageLine, (160, 90), (220, 85), (0, 255, 0), thickness=2, lineType=cv2.LINE_AA)
display_image(imageLine)
```

```python
imageCircle = image.copy()
cv2.circle(imageCircle, (190, 90), 80, (0, 255, 0), thickness=2, lineType=cv2.LINE_AA)
display_image(imageCircle)
```

```python
imageEllipse = image.copy()
cv2.ellipse(imageEllipse, (200, 310), (50, 80), 45, 0, 360, (255, 0, 0), thickness=2, lineType=cv2.LINE_AA)
cv2.ellipse(imageEllipse, (200, 310), (50, 80), 135, 0, 360, (0, 0, 255), thickness=2, lineType=cv2.LINE_AA)
display_image(imageEllipse)
```

```python
imageRectangle = image.copy()
cv2.rectangle(imageRectangle, (130, 5), (250, 180), (0, 255, 0), thickness=2, lineType=cv2.LINE_8)
display_image(imageRectangle)
```

```python
imageText = image.copy()
cv2.putText(imageText, "Alejandro E. Rendon", (50, 200), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
display_image(imageText)
```
