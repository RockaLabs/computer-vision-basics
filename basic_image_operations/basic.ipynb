# Basic Image Operations
## Read, Write and Display

```python
import cv2
import matplotlib.pyplot as plt
```

Load imagen and check if exist

```python
image = cv2.imread("rocka.png")
photo = cv2.imread("photo.jpg")

if image is None:  # Check for invalid input
    print("Could not open or find the image")
    
if photo is None:  # Check for invalid input
    print("Could not open or find the image")
```

```python
plt.imshow(image)
plt.show()
```

The image has been correctly loaded by openCV as a numpy array, but the color of each pixel has been sorted as `BGR`.
Matplotlib's plot expects an `RGB` image so, for a correct display of the image, it is necessary to swap those channels. 
This operation can be done either by using openCV conversion functions `cv2.cvtColor()` or by working directly with the numpy array.

```python
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
photo = cv2.cvtColor(photo, cv2.COLOR_BGR2RGB)
```

Display image

```python
plt.imshow(image)
plt.show()
```

```python
plt.imshow(photo)
plt.show()
```

## Crop and Resize

### Resize

The resize function has the following prototype:

`destinationImage = cv2.resize(sourceImage, (,), scale factor in X direction, scale factor in Y direction, interpolation method)`


```python
source = load_image('rocka.png')

scaleX = 0.5
scaleY = 0.8
```

```python
scaleDown = cv2.resize(source, None, fx= scaleX, fy= scaleY, interpolation= cv2.INTER_LINEAR)
display_image(scaleDown)
```

```python
scaleUp = cv2.resize(source, None, fx= scaleX*2, fy= scaleY*3, interpolation= cv2.INTER_LINEAR)
display_image(scaleUp)
```

### Crop

`Array[rows, columns]`

```python
crop = source[100:300, 100:300]
display_image(crop)
```

## Rotation
### Matrix Rotation

```python
source = load_image('rocka.png')

dim = source.shape

rotationAngle = 30
scaleFactor = 1
```

```python
rotationMatrix = cv2.getRotationMatrix2D((dim[1]/2, dim[0]/2), rotationAngle, scaleFactor)
result = cv2.warpAffine(source, rotationMatrix, (dim[1],dim[0]))
```

```python
display_image(result)
```
